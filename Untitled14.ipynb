```python
#intersection between two list
A = {1,2,3,4,5}
B = {4,5,6,7,8}
A.intersection(B)
```

```python
# union function
A.union(B)
```

```python
# difference function
A.difference(B)
```

```python
B.difference(A)
```

```python
(A.difference(B)).union(B.difference(A))
```

```python
#symmetric difference
A={1,2,3,4}
B={3,4,5,6}
print(A.symmetric_difference(B))
```

```python
#issuperset
A={1,2,3,4}
B={3,4,5,6}
print(A.issuperset(B))
```

```python
#isdisjoint
A={1,2,3,4}
B={7,8,5,6}
print(A.isdisjoint(B))
```
