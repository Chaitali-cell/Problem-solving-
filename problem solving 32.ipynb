```python
1.#Filter Palindrome Words
words = ["madam", "python", "level", "hello", "radar"]
result = list(filter(lambda x: x == x[::-1], words))
print(result)
```

```python
2.#Filter Prime Numbers
def prime(num):
    if num < 2:
        return False
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True
numbers = [2, 3, 4, 5, 6, 7, 8, 9, 10]
result = list(filter(prime, numbers))
print(result)
```

```python
3.#Filter Students Passed (Marks ≥ 40)
marks = [35, 78, 90, 20, 55, 39]
passed = list(filter(lambda x: x >= 40, marks))
print(passed)
```

```python
4.#Filter Employees with Salary Above 50,000
salary = [25000, 60000, 45000, 80000, 55000]
result = list(filter(lambda x: x > 50000, salary))
print(result)
```

```python
5.#Filter Leap Years
years = [2000, 2001, 2004, 2021, 2024]
result = list(filter(lambda y: (y % 400 == 0) or (y % 4 == 0 and y % 100 != 0), years))
print(result)
```

```python
6.#Filter Dictionary (Age > 18)
students = [
    {"name": "Aman", "age": 17},
    {"name": "Riya", "age": 20},
    {"name": "Rahul", "age": 19}
]
result = list(filter(lambda x: x["age"] > 18, students))
print(result)
```

```python
7.#Filter Numbers Divisible by Both 3 and 5
numbers = list(range(1, 51))
result = list(filter(lambda x: x % 3 == 0 and x % 5 == 0, numbers))
print(result)
```

```python
8.#Filter Valid Email Addresses
emails = [
    "abc@gmail.com",
    "python",
    "test@yahoo.com",
    "hello"
]
result = list(filter(lambda x: "@" in x and "." in x, emails))
print(result)
```

```python
9.#Filter Perfect Squares
import math
numbers = [4, 7, 9, 10, 16, 20, 25]
result = list(filter(lambda x: math.sqrt(x).is_integer(), numbers))
print(result)
```

```python
10.#Filter Numbers Between 20 and 50
numbers = [5, 15, 22, 35, 48, 60, 75]
result = list(filter(lambda x: 20 <= x <= 50, numbers))
print(result)
```
