```python
1.#Find Longest Word
from functools import reduce
words = ["Apple", "Watermelon", "Cat", "Banana"]
longest = reduce(lambda x, y: x if len(x) > len(y) else y, words)
print(longest)
```

```python
2.#Multiply All Odd Numbers
from functools import reduce
numbers = [1, 2, 3, 4, 5]
odd = list(filter(lambda x: x % 2 != 0, numbers))
result = reduce(lambda x, y: x * y, odd)
print(result)
```

```python
3.#Partial Function for GST
from functools import partial
def bill(amount, gst):
    return amount + amount * gst / 100
gst18 = partial(bill, gst=18)
print(gst18(500))
```

```python
4.#Find Highest Salary
from functools import reduce
salary = [25000, 30000, 28000, 35000]
highest = reduce(lambda x, y: x if x > y else y, salary)
print(highest)
```

```python
5.#Sum of Squares
from functools import reduce
numbers = [1, 2, 3, 4]
result = reduce(lambda x, y: x + y**2, numbers, 0)
print(result)
```

```python
6.#Count Vowels
from functools import reduce
text = "DataAnalyst"
count = reduce(lambda x, y: x + 1 if y.lower() in "aeiou" else x, text, 0)
print(count)
```

```python
7.#Count Consonants
from functools import reduce
text = "Python"
count = reduce(
    lambda x, y: x + 1 if y.isalpha() and y.lower() not in "aeiou" else x,
    text,
    0
)
print(count)
```

```python
8.#Find Total Length of Words
from functools import reduce
words = ["Python", "SQL", "Excel"]
length = reduce(lambda x, y: x + len(y), words, 0)
print(length)
```

```python
9.#Count Positive Numbers
from functools import reduce

numbers = [-2, 5, -7, 8, 10]

count = reduce(lambda x, y: x + 1 if y > 0 else x, numbers, 0)

print(count)
```

```python
10.#Count Negative Numbers
from functools import reduce
numbers = [-2, 5, -7, 8, 10]
count = reduce(lambda x, y: x + 1 if y < 0 else x, numbers, 0)
print(count)
```
