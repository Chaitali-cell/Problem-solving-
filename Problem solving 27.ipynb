```python
1.#Salary Slip Generator
def salary_slip(basic):
    hra = basic * 0.20
    da = basic * 0.10
    pf = basic * 0.12
    net_salary = basic + hra + da - pf
    return net_salary

basic = float(input("Enter Basic Salary: "))
print("Net Salary = ₹", salary_slip(basic))
```

```python
2.#Sales Commission
def commission(sales):
    return sales * 0.05
sales = float(input("Enter Sales Amount: "))
print("Commission = ₹", commission(sales))
```

```python
3.#Bonus Calculator
def bonus(salary):
    return salary * 0.15

salary = float(input("Enter Salary: "))
print("Bonus = ₹", bonus(salary))
```

```python
4.#Internet Bill
def internet(months):
    return months * 600
months = int(input("Enter Number of Months: "))
print("Internet Bill = ₹", internet(months))
```

```python
5.#Car Rental Bill
def car_rent(days):
    return days * 1800

days = int(input("Rental Days: "))
print("Total Rent = ₹", car_rent(days))
```

```python
6.#Hotel Booking Discount
def hotel_bill(amount):
    if amount > 10000:
        amount -= amount * 0.20
    return amount
bill = float(input("Hotel Bill: "))
print("Final Bill = ₹", hotel_bill(bill))
```

```python
7.#Fruit Shop Bill
def fruit_bill(apples, mangoes):
    return (apples * 120) + (mangoes * 90)
a = int(input("Apples (kg): "))
m = int(input("Mangoes (kg): "))
print("Bill = ₹", fruit_bill(a, m))
```

```python
8.#Employee Attendance
def attendance(total_days, present_days):
    return (present_days / total_days) * 100
total = int(input("Working Days: "))
present = int(input("Present Days: "))
print("Attendance =", attendance(total, present), "%")
```

```python
9.#Overtime Salary
def overtime(hours):
    return hours * 250
hours = int(input("Overtime Hours: "))
print("Extra Salary = ₹", overtime(hours))
```

```python
10.#Printing Shop
def printing(pages):
    return pages * 3

pages = int(input("Pages: "))
print("Total Cost = ₹", printing(pages))
```
