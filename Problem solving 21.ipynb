```python
#wap to print sum of even number in a list
import functools
numbers = [1, 2, 3, 4, 5, 6]
functools.reduce(lambda x, y: x + y, filter(lambda n: n % 2 == 0, numbers))
```

```python
#To print largest word in a list
from functools import reduce
words = ["cat", "elephant", "dog", "tiger"]
longest = reduce(lambda x, y: x if len(x) > len(y) else y, words)
print(longest)
```

```python
#To print smallest word in a list
from functools import reduce
words = ["cat", "elephant", "dog", "tiger"]
shortest = reduce(lambda x, y: x if len(x) < len(y) else y, words)
print(shortest)
```
