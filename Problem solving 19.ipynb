```python
#Addition of two number using lambda function
Res=lambda a,b:a+b
Res(5,2)
```

```python
#Subtraction of two number using lambda function
Res=lambda a,b:a-b
Res(5,2)
```

```python
#Multiplication of two number using lambda function
Res=lambda a,b:a*b
Res(5,2)
```

```python
#Divison of two number using lambda function
Res=lambda a,b:a/b
Res(10,2)
```

```python
#Remainder of two number using lambda function
Res=lambda a,b:a%b
Res(10,2)
```

```python
#Find square root of number using lambda function
sqrt=lambda x:x**(1/2)
sqrt(25)
```

```python
#Find cube root of number using lambda function
Cbrt=lambda y:y**(1/3)
Cbrt(81)
```

```python
#To find area of rectangle using lambda function
ra=lambda l,b,h:l*b*h
ra(5,2,2)
```

```python
#PRINT GREATER NUMBER BETWEEN TWO NUMBER USING LAMBDA
gn=lambda a,b: a if a>b else b
gn(30,20)
```

```python
#To find area of circle using lambda function
CA=lambda r:3.14*r**2
CA(2.5)
```
