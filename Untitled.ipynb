```python
#find square root of number
import math
n = int(input("Enter number"))
print("Square root",math.sqrt(n))
```

```python
#find cube root of number
import math
n = int(input("Enter number"))
print("cube root",math.cbrt(n))
```

```python
#find power
import math
base = int(input("Enter base: "))
exp = int(input("Enter exponent: "))
print("Result:", math.pow(base, exp))
```

```python
#find factorial
import math
num = int(input("Enter a number: "))
print("Factorial:", math.factorial(num))
```

```python
#find ceiling and floor value
import math
num = float(input("Enter a decimal number: "))
print("Ceiling:", math.ceil(num))
print("Floor:", math.floor(num))
```

```python
#find absolute value
import math
num = int(input("Enter a number: "))
print("Absolute Value:", math.fabs(num))
```

```python
#find lcm
import math
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
print("LCM:", math.lcm(a, b))
```

```python
#find log
import math
num = int(input("Enter a number: "))
print("Natural Log:", math.log(num))
```

```python
#find sin,cos,tan
import math
angle = int(input("Enter angle in degrees: "))
radian = math.radians(angle)
print("Sin:", math.sin(radian))
print("Cos:", math.cos(radian))
print("Tan:", math.tan(radian))
```

```python
#find distance
import math
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
print("Distance:", math.fabs(a - b))
```
