```python
#WAP to print vowels in words
word=(input("Enter Word"))
for vowel in word:
    if vowel=='a' or vowel=='e' or vowel=='i' or vowel=='o' or vowel=='u':
        print(vowel)
```

```python
#WAP TO CALCULATE FACTORIAL OF GIVEN USER NUMBER
num1=int(input("Enter number"))
res=1
i=1
while(i<=num1):
    res=res*i
    i=i+1
print(res)
```

```python
#wap on multiplication table for user given number
a=int(input("Enter number"))
for i in range(1,11):
    print(i*a)
```

```python
#wap on multiplication table from 11 to 20
for i in range(11,21):
    for j in range(1,11):
        print(j*i,end=' ')
    print()
```
