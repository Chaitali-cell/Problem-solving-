```python
#To print even odd using filter 
l1 = [2,3,1,4,6,5,7]
def EV(x):
    if x%2==0:
        return True
    else:
        return False
filter(EV,l1)    
```

```python
list(filter(EV,l1))
```

```python
list(filter(lambda x: x%2==0,l1))
```

```python
#add two numbers
add = lambda a,b: a+b
print(add(10,20))
```

```python
#sub two numbers
sub = lambda a,b: a-b
print(sub(20,10))
```

```python
#check even or odd number
EVOD = lambda x: "EVEN" if x%2==0 else "ODD"
print(EVOD(25))
```

```python
#find the minimum number using lambda function
MIN = lambda a,b:a if a<b else b
print(MIN(30,40))
```

```python
#find the maximum number using the lambda function
MAX= lambda a,b : a if a>b else b
print(MAX(30,40))
```

```python
#find the largest of three number
LAR = lambda a,b,c: max(a,b,c)
print("The largest number is:",LAR(30,20,90))
```

```python
#find the smallest of three number
SMALL=lambda a,b,c: min(a,b,c)
print("The smallest number is",SMALL(30,20,80))
```

```python
#convert the uppercase using lambda function 
upper = lambda s:s.upper()
print(upper("python"))
```

```python
#convert the lowercase using lambda function
lower = lambda s:s.lower()
print(lower("PYTHON"))
```
