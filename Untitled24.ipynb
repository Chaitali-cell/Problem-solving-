```python
#create a function to calculate electric bill
def electricity_bill(unit):
    if unit <=100:
        bill=unit*5
    elif unit<=200:
        bill=unit*7
    else:
        bill=unit*10
    return bill
print("bill:",electricity_bill(250))
        
```

```python
#salary calender
def salary(basic):
    hra = basic * 0.20
    da = basic * 0.10
    
    total = basic + hra + da
    
    return total
print("Salary =", salary(30000))
```

```python
#Shopping Discount
def discount(amount):
    if amount >= 5000:
        return amount * 0.20
    elif amount >= 2000:
        return amount * 0.10
    else:
        return 0

print("Discount =", discount(6000))
```

```python
#Bank Interest Calculator
def interest(p, r, t):
    return (p*r*t)/100

print(interest(10000,5,2))
```

```python
#Age Calculator
def age(birth_year):
    current_year = 2026
    return current_year - birth_year
print("Age =", age(2004))
```

```python
#Taxi Fare Calculator
def taxi_fare(distance):
    fare = distance * 15
    return fare
print("Fare =", taxi_fare(10))
```

```python
#ATM Withdrawal
def withdraw(balance, amount):
    if amount <= balance:
        return balance - amount
    else:
        return "Insufficient Balance"

print(withdraw(10000,3000))
```

```python
#BMI checker
def bmi(weight, height):
    result = weight/(height**2)
    
    return result

print(bmi(70,1.75))
```

```python
#Employee Bonus Calculator
def bonus(salary):
    if salary > 50000:
        return salary*0.10
    else:
        return salary*0.05

print(bonus(60000))
```

```python
# Voting Eligibility
def can_vote(age):
    if age >= 18:
        return "Eligible"
    else:
        return "Not Eligible"

print(can_vote(19))
```
