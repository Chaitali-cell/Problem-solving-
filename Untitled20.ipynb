```python
#To print area of square using def function
def area_of_square(side):
    s = (side)**2
    return s

total = area_of_square(4)
print(total)
```

```python
#To print area of rectangle using def function
def area_of_rect(l,b):
    s = l*b
    return s

total = area_of_rect(4,2)
print(total)
```

```python
#To print area of triangle using def function
def area_of_tri(b,h):
    s = 1/2*b*h
    return s

total = area_of_tri(3,4)
print(total)
    
```

```python
#To print area of circle using def function
def area_of_circle(r):
    s = 3.14*r**2
    return s

total = area_of_circle(2)
print(total)
```

```python
#To find the largest number using the def function
def largest_num(a,b):
    if a>b:
        print("a is largest number")
        return a
    else:
        print("b is largest number")
        return b

result = largest_num(3,5)
print(result)
```

```python
#To find the smallest number using the def function
def smallest_num(a,b):
    if a<b:
        print("a is smallest number")
        return a
    else:
        print("b is smallest number")
        return b

result = smallest_num(3,5)
print(result)
```

```python
# To find the even or odd number using def function
def even_odd(num):
    if num%2==0:
        print(f"{num} is even number")
    else:
        print(f"{num} is odd number")

result = even_odd(10)
print(result)
```

```python
# To print factorial of a number using def function
def factorial(n):
    fact = 1
    for i in range(1,n+1):
        fact = fact * i
    return fact

result = factorial(4)
print(result)
```

```python
#Find maximum number in a list
def find_max(numbers):
    maximum = numbers[0]
    for i in numbers:
        if i > maximum:
            maximum = i
    return maximum

nums = [10, 45, 22, 67, 34]
print("Maximum =", find_max(nums))
```

```python
#Reverse the string
def reverse_string(text):
    return text[::-1]

print(reverse_string("Python"))
```
