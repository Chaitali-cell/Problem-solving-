```python
1.#Find the Smaller Number
small = lambda a, b: a if a < b else b
print(small(25, 10))
```

```python
5.#Check Divisible by 5
check = lambda n: "Yes" if n % 5 == 0 else "No"
print(check(35))
```

```python
3.#Calculate Discount
discount = lambda price, dis: price - (price * dis / 100)
print(discount(2000, 10))
```

```python
4.#Calculate GST (18%)
gst = lambda amount: amount + amount * 18 / 100

print(gst(500))
```

```python
5.#Convert String to Uppercase
upper = lambda text: text.upper()
print(upper("python"))
```

```python
6.#Convert String to Lowercase
lower = lambda text: text.lower()
print(lower("PYTHON"))
```

```python
7.#Check String Starts with 'P'
check = lambda text: text.startswith("P")
print(check("Python"))
```

```python
8.#Convert List to Absolute Values
numbers = [-5, -3, 4, -7]
result = list(map(lambda x: abs(x), numbers))
print(result)
```

```python
9.#Sort Dictionary by Value
marks = {"Ram": 80, "Amit": 95, "Riya": 75}
result = sorted(marks.items(), key=lambda x: x[1])
print(result)
```

```python
10.#Sort Strings by Length
words = ["apple", "kiwi", "banana", "mango"]
result = sorted(words, key=lambda x: len(x))
print(result)
```
