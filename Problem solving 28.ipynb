```python
1.#Student Scholarship
def scholarship(percentage, attendance):
    if percentage >= 85 and attendance >= 90:
        return "Scholarship Approved"
    else:
        return "Scholarship Not Approved"

percentage = float(input("Enter Percentage: "))
attendance = float(input("Enter Attendance: "))

print(scholarship(percentage, attendance))
```

```python
2.#Monthly Savings
def savings(income, expenses):
    return income - expenses

income = float(input("Enter Monthly Income: "))
expenses = float(input("Enter Monthly Expenses: "))

print("Savings:", savings(income, expenses))
```

```python
3.#Product Stock Alert
def stock_alert(stock):
    if stock < 20:
        return "Restock Required"
    else:
        return "Stock Available"

stock = int(input("Enter Stock Quantity: "))
print(stock_alert(stock))
```

```python
4.#Calculate Income Tax
def income_tax(salary):
    if salary > 800000:
        tax = salary * 0.20
    elif salary > 500000:
        tax = salary * 0.10
    else:
        tax = 0
    return tax

salary = float(input("Enter Annual Salary: "))
print("Income Tax:", income_tax(salary))
```

```python
5.#Employee Overtime Pay
def overtime_pay(hours):
    if hours > 40:
        overtime = (hours - 40) * 300
    else:
        overtime = 0

    return overtime

hours = int(input("Enter Working Hours: "))
print("Overtime Pay:", overtime_pay(hours))
```

```python
6.#Find Average Age
def average_age(ages):
    return sum(ages) / len(ages)

ages = [21, 24, 22, 25, 23]

print("Average Age:", average_age(ages))
```

```python
7.#Check Strong Password
def check_password(password):

    if len(password) >= 8:
        return "Strong Password"
    else:
        return "Weak Password"

password = input("Enter Password: ")

print(check_password(password))
```

```python
8.#Check Perfect Square
def perfect_square(num):
    root = int(num ** 0.5)

    if root * root == num:
        return "Perfect Square"
    else:
        return "Not Perfect Square"

num = int(input("Enter Number: "))

print(perfect_square(num))
```

```python
9.#Find the Second Smallest Number
def second_smallest(numbers):

    first = second = float('inf')

    for num in numbers:
        if num < first:
            second = first
            first = num
        elif num < second and num != first:
            second = num

    return second

numbers = [25, 12, 45, 8, 19, 30]

print("Second Smallest:", second_smallest(numbers))
```

```python
10.#Count Employees Eligible for Promotion
def promotion(experience):

    count = 0

    for year in experience:
        if year >= 5:
            count += 1

    return count

experience = [2, 6, 8, 3, 5, 10]

print("Eligible:", promotion(experience))
```
