```python
#WAP on pattern
for i in range(1,5):
    if(i%2==0):
        print("#"*i)
    else:
        print("&"*i)
```

```python
#WAP using break keyword
for i in range(1,11):
    if i==5:
        break
    print(i)
else:
    print("hello")
```

```python
#WAP using continue keyword
for i in range(1,11):
    if i==5 or i==7 or i==9:
        continue
    print(i)
else:
    print("hello")
```

```python
#WAP using pass keyword
for i in range(1,11):
    if i==5:
        pass
    print(i)
else:
    print("hello")
```

```python
#WAP to check user given number is prime or not?
num=int(input("Enter Number"))
for i in range(2,num):
    if(num%i==0):
        print("Number is not prime")
        break
else:
    print("Number is prime")
```

```python
#wap on incrementing number triangle
num=int(input("Enter number"))
for i in range(1,num+1):
    for j in range(1,i+1):
        print(j,end=' ')
    print()
```

```python
#wap on decrementing number triangle
rows=int(input("Enter Number"))
for i in range(rows, 0, -1):
    for j in range(1, i + 1):
        print(j, end=" ")
    print()
```

```python
#wap on pyramid pattern
rows = 5
for i in range(1, rows + 1):
    print(" " * (rows - i) + "*" * (2 * i - 1))
```

```python
rows = 5
for i in range(rows, 0, -1):
    print(" " * (rows - i) + "*" * (2 * i - 1))
```
