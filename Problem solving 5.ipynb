```python
#WAP on diamond "*" pattern
rows = 5
for i in range(1, rows + 1):
    print(" " * (rows - i) + "*" * (2 * i - 1))
for i in range(rows, 0, -1):
    print(" " * (rows - i) + "*" * (2 * i - 1))
```

```python
#WAP on "*" aligned inverted triangle pattern program 
rows=5
for i in range(5,0,-1):
    print(" "*(rows-i)+"*"*i)
```

```python
#WAP on "*" square pattern program 
rows=5
for i in range(5):
    for j in range(5):
        print("*",end=" ")
    print()
```

```python
#WAP on number square pattern program 
rows=5
for i in range(1,5):
    for j in range(1,5):
        print(j,end=" ")
    print()
```

```python
#WAP ON HOLLOW SQUARE PATTERN
rows=int(input("Enter number"))
for i in range(rows):
    for j in range(rows):
        if i==0 or i==rows-1 or j==0 or j==rows-1:
            print("*",end=" ")
        else:
            print(" ", end=" ")
    print()
         
```
