```python
1.#ATM Withdrawal

balance = float(input("Enter account balance: "))
withdraw = float(input("Enter withdrawal amount: "))
if withdraw <= balance:
    balance = balance - withdraw
    print("Transaction Successful")
    print("Remaining Balance =", balance)
else:
    print("Insufficient Balance")
```

```python
2.#Library Fine

days = int(input("Enter late days: "))
if days <= 5:
    fine = days * 2
elif days <= 10:
    fine = (5 * 2) + ((days - 5) * 5)
else:
    fine = (5 * 2) + (5 * 5) + ((days - 10) * 10)

print("Library Fine =", fine)
```

```python
3.#Petrol Expense Calculator

distance = float(input("Enter distance (km): "))
mileage = float(input("Enter mileage (km/litre): "))
petrol_price = float(input("Enter petrol price per litre: "))
litres = distance / mileage
cost = litres * petrol_price
print("Fuel Required =", litres, "litres")
print("Total Cost =", cost)
```

```python
4.#Movie Ticket Booking

tickets = int(input("Enter number of tickets: "))
price = tickets * 200
if tickets > 5:
    discount = price * 0.10
else:
    discount = 0
final_price = price - discount
print("Total Ticket Price =", final_price)
```

```python
5.#Hotel Room Rent

days = int(input("Enter number of days: "))
rent = days * 1500
if days > 10:
    discount = rent * 0.25
elif days > 5:
    discount = rent * 0.15
else:
    discount = 0

final_rent = rent - discount

print("Room Rent =", rent)
print("Discount =", discount)
print("Final Amount =", final_rent)
```

```python
6.#Parking Fee Calculator

hours = int(input("Enter parking hours: "))

if hours <= 2:
    fee = hours * 20
else:
    fee = (2 * 20) + ((hours - 2) * 30)

print("Parking Fee =", fee)
```

```python
7.#Hospital Bill

days = int(input("Enter number of days admitted: "))
charge = float(input("Enter charge per day: "))

bill = days * charge

print("Hospital Bill =", bill)
```

```python
8.#Loan EMI (Simple)
loan = float(input("Enter loan amount: "))
months = int(input("Enter number of months: "))

emi = loan / months

print("Monthly EMI =", emi)
```

```python
9.#Mobile Recharge Cashback
amount = float(input("Enter recharge amount: "))
if amount >= 500:
    cashback = amount * 0.10
elif amount >= 300:
    cashback = amount * 0.05
else:
    cashback = 0

print("Cashback =", cashback)
print("Final Amount Paid =", amount - cashback)
```

```python
10.#Internet Data Usage
used = float(input("Enter data used (GB): "))
limit = 100

if used > limit:
    print("Data Limit Exceeded")
else:
    print("Remaining Data =", limit - used, "GB")
```
