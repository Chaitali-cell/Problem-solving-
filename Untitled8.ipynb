```python
#find the largest element in the list
numbers = [10, 45, 2, 78, 23]
largest = numbers[0]
for num in numbers:
    if num > largest:
        largest = num

print("Largest number:", largest)
        
```

```python
#find the smallest element in the list
numbers = [12,45,3,34,7,54]
smallest = numbers[0]
for num in numbers:
    if num<smallest:
        smallest = num

print("smallest number:",smallest)
```

```python
#find sum of all element in a list
numbers = [10, 20, 30, 40]
total = 0
for num in numbers:
    total += num
print("Sum =", total)
```

```python
#count odd and even in a list
numbers = [10, 15, 22, 7, 8, 13]
even = 0
odd = 0
for num in numbers:
    if num % 2 == 0:
        even += 1
    else:
        odd += 1

print("Even:", even)
print("Odd:", odd)
```

```python
#removes duplicate from a list
numbers = [3,2,54,4,3,2,5,3]
unique = []
for num in numbers:
    if num not in unique:
        unique.append(num)

print(unique)
```

```python
#count occurences of an element in a list
t = [1, 2, 3, 2, 4, 2, 5]
count = 0
for num in t:
    if num == 2:
        count += 1

print("Count:", count)
```
