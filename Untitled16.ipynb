```python
#check if key is exist
student = {"Name":"Zebra","Age":20}
if "Age" in student:
    print("key found")
else:
    print("key not found")
```

```python
#copy a dictionary
d1 = {"A":10, "B":20}
d2 = d1.copy()
print(d2)
```

```python
#clear a dictionary
d1 = {"A":"CHAITALI", "B":20}
d1.clear()
print(d1)
```

```python
#Remove key using pop
d1 = {"A":"CHAITALI", "B":20}
d1.pop("A")
print(d1)
```

```python
#update two dictionary
d1= {"A":"CHAITALI", "B":20}
d2 = {"A":"SANJANA", "B":22}
d1.update(d2)
print(d1)
```

```python
#print keys 
d1= {"A":"CHAITALI", "B":20}
for key in d1:
    print(key)
```

```python
#print value
d1= {"A":"CHAITALI", "B":20}
for value in d1.values():
    print(value)
```

```python
#print key value
d1= {"A":"CHAITALI", "B":20}
for key,value in d1.items():
    print(key,":",value)
```

```python
#create a dictionary using list
d1 = ["Name","Age","city"]
d2 = ["chaitali",22,"pune"]
S=(dict(zip(d1,d2)))
print(S)
```
