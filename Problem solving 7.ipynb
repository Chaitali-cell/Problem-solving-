```python
#find lcm of two number
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

max_num = max(a, b)

while True:
    if max_num % a == 0 and max_num % b == 0:
        print("LCM =", max_num)
        break
    max_num += 1
```

```python
#wap on floyd's triange
rows = 5
num = 1

for i in range(1, rows + 1):
    for j in range(i):
        print(num, end=" ")
        num += 1
    print()
    
```

```python
#wap to sort a list in ascending order
numbers = [50, 10, 30, 20, 40]
numbers.sort()
print(numbers)
```

```python
#wap to sort a list in descending order
numbers = [50, 10, 30, 20, 40]

numbers.sort(reverse=True)

print(numbers)
```

```python
#wap on plus star pattern
rows = 5
for i in range(rows):
    for j in range(rows):
        if i == rows // 2 or j == rows // 2:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
```
