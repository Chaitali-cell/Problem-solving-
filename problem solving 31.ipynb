```python
1.#Filter Even Numbers
numbers = [10, 15, 20, 25, 30, 35]
result = list(filter(lambda x: x % 2 == 0, numbers))
print(result)
```

```python
2.#Filter Odd Numbers
numbers = [10, 15, 20, 25, 30, 35]
result = list(filter(lambda x: x % 2 != 0, numbers))
print(result)
```

```python
3.#Filter Positive Numbers
numbers = [-5, 10, -2, 8, -9, 15]
positive = list(filter(lambda x: x > 0, numbers))
print(positive)
```

```python
4.#Filter Negative Numbers
numbers = [-5, 10, -2, 8, -9, 15]
negative = list(filter(lambda x: x < 0, numbers))
print(negative)
```

```python
5.#Filter Names Starting with 'A'
names = ["Aman", "Rahul", "Anjali", "Riya", "Akash"]
result = list(filter(lambda name: name.startswith("A"), names))
print(result)
```

```python
6.#Filter Strings Longer Than 5 Characters
words = ["cat", "elephant", "dog", "computer", "pen"]
result = list(filter(lambda word: len(word) > 5, words))
print(result)
```

```python
7.#Filter Numbers Greater Than 50
numbers = [10, 55, 78, 20, 99, 45]
result = list(filter(lambda x: x > 50, numbers))
print(result)
```

```python
8.#Filter Multiples of 5
numbers = [11, 25, 30, 18, 40, 22]
result = list(filter(lambda x: x % 5 == 0, numbers))
print(result)
```

```python
9.#.filter Vowels from Characters
letters = ['a', 'b', 'e', 'g', 'i', 'o', 'u', 'x']
vowels = list(filter(lambda x: x in "aeiou", letters))
print(vowels)
```

```python
10.#Filter Empty Strings
words = ["Python", "", "Java", "", "SQL", ""]
result = list(filter(None, words))
print(result)
```
