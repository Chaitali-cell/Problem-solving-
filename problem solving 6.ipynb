```python
#WAP TO CHECK LEAP YEAR
year = int(input("Enter year: "))

if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print("Leap Year")
else:
    print("Not a Leap Year")
```

```python
#WAP ON SUM OF FIRST N NATURAL NUMBERS
n = int(input("Enter N: "))
sum = 0

for i in range(1, n + 1):
    sum += i

print("Sum =", sum)
```

```python
#wap to check palindrone number
num = int(input("Enter a number: "))
temp = num
rev = 0

while num > 0:
    digit = num % 10
    rev = rev * 10 + digit
    num //= 10

if temp == rev:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
#Wap to check Armstrong number
num = int(input("Enter a number: "))
temp = num
sum = 0

while num > 0:
    digit = num % 10
    sum += digit ** 3
    num //= 10

if temp == sum:
    print("Armstrong Number")
else:
    print("Not Armstrong Number")
```

```python
#wap to count digit in a number
num = int(input("Enter a number: "))
count = 0

while num > 0:
    count += 1
    num //= 10

print("Number of digits:", count)
```

```python
#wap to find sum of digit
num = int(input("Enter a number: "))
sum = 0

while num > 0:
    digit = num % 10
    sum += digit
    num //= 10

print("Sum of digits =", sum)
```

```python
#wap on swapping two number
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

a, b = b, a

print("After swapping:")
print("a =", a)
print("b =", b)
```

```python
#wap to find positive,negative or zero
num = int(input("Enter a number: "))

if num > 0:
    print("Positive")
elif num < 0:
    print("Negative")
else:
    print("Zero")
```

```python
#wap to find largest among three value
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

if a >= b and a >= c:
    print("Largest =", a)
elif b >= a and b >= c:
    print("Largest =", b)
else:
    print("Largest =", c)
```
