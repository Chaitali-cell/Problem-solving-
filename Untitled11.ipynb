```python
#WAP TO SORT THE LIST
l1=[10,45,65,70, 20, 30]
sorted(l1)
```

```python
#WAP TO SORT THE LIST IN REVERSE ORDER
sorted(l1,reverse=True)
```

```python
#Wap to clear the list
l1.clear()
l1
```

```python
#WAP to find all true or false
l2=[10,45,65,70, 20, 30]
all(l2)
```

```python
l3 = [10,20,30,40,50,0]
all(l3)
```
