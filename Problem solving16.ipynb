```python
#reverse every word
s = input("Enter the string")
words = s.split()
for word in words:
    print(word[::-1],end =" ")
```

```python
#count frequency of each character
s = input("Enter the string : ")
for ch in set(s):
    print(ch,"=",s.count(ch))
```

```python
#remove the duplicate character
s = input("Enter the string")
result = ""
for ch in s:
    if ch not in result:
        result+=ch
print("new string",result)
```

```python
#count uppercase and lowercae letters
s = input("enter a string")
upper = lower = 0
for ch in s:
    if ch.isupper():
        upper+=1
    elif ch.islower():
        lower+=1
print("uppercase = ",upper)
print("lowercase = ",lower)
```
